```python
def strategy(start_money, predictions, real, sell_coef, buy_count=1):
    current = 0;
    history = []
    
    upper_bound = predictions.max()
    lower_bound = predictions.min()
    
    for i in real:
        history.append(start_money)
        if start_money <=0:
            return start_money, history
        
        if i < lower_bound:
            current += buy_count
            start_money -= i * buy_count
        elif i > upper_bound:
            current -= int(current * sell_coef)
            start_money += int(current * sell_coef) * i
    start_money += current * i
    
    return (start_money, history)
```

```python
coef = np.arange(0.01, 1.01, 0.01)
```

```python
bc = np.arange(1, 100, 1)
```

```python
history = []
for j in coef:
    for k in bc:
        history.append(strategy(10000, pred, y_test, j, k)[0])
```
